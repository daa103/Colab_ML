##**04-2 matplotlib 라이브러리 자유자재로 사용하기**

> ### 기초그래프 그리기-히스토그램, 산점도 그래프, 박스 그래프

```python
tips=sns.load_dataset('tips')
print(tips.head())
print(type(tips))
```

>> 2. 히스토그램은 데이터프레임의 열 데이터 분포와 빈도를 살피는 용도의 그래프프, total_bill, tip등의 열을 변수라고 부른다. 변수를 하나만 사용하는것은 '일변량 그래프'이다.
>> 3. 기본틀(fig)마련하고 격자(axes1)를 구성


```python
fig=plt.figure()
axes1=fig.add_subplot(1,1,1)
```

>> 4. hist메서드에 total_bill열을 전달하면 히스토그램이 만들어진다. x축의 간격은 bins인잣값으로 조정.

```python
axes1.hist(tips['total_bill'],bins=10)
axes1.set_title('Histogram of Total Bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')
fig
```

>> 5. 산점도는 변수 2개를 사용해서 만든 것은 '이변량 그래프'이다.
>> 6. 기본틀과 그래프 격자를 만들고 scatter메서드에 total_bill, tips 열을 전달하면  total_bill과 tips에 대한 산점도 그래프를 그린다.

```python
scatter_plot=plt.figure()
axes1=scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'],tips['tip'])
axes1.set_title('Scatterplot of Totla Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

>> 7. 박스그래프는 이산형변수와 연속형 변수를 함께사용한다.이산형변수는 명확하게 구분되는 값이고 연속형 변수는 셀수없는 범위를 의미한다.
>> 8. boxplot메서드를 이용해 박스 그래프를 그린다. tips 데이터프레임에서 labels인잣값으로 성별을 구분하여 리스트를 작성하다.

```python
boxplot=plt.figure()
axes1=boxplot.add_subplot(1,1,1)
axes1.boxplot([tips[tips['sex']=='Female']['tip'],
            tips[tips['sex']=='Male']['tip']],
           labels=['Female','Male'])
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
```

> ### 다변량 그래프 그리기
* 3개이상 변수를 사용한 그래프는 다변량 그래프
> #### 다변량 데이터로 다변량 그래프 그리기-산점도 그래프
>> 1. scatter_plot에 성별을 새 변수로 추가할때, 문자열은 산점도 그래프의  색상을 지정하는 값으로 사용할 수 없기 때문에 치환하는 함수를 만들어야한다.

```python
def recode_sex(sex):
  if sex=='Female':
    return 0
  else:
    return 1
```

>> 2. recode_sex가 반환한 값을 데이터프레임에 추가하면 데이터프레임에 새로운 열을 추가한다.

```python
tips['sex_color']=tips['sex'].apply(recode_sex)
```

>> 3. 인원수도 치환값을 구하여 산점도에 추가한다.

```python
scatter_plot=plt.figure()
axes1=scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x=tips['total_bill'],
    y=tips['tip'],
    s=tips['size']*10,
    c=tips['sex_color'],
    alpha=0.5)
axes1.set_title('Total Bill vs Tip colored by sex and sized by size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

> ## **seaborn 라이브러리로 그래프 스타일 설정**

>> whitegrid스타일을 설정한다다

```python
sns.set_style('whitegrid')
fig,ax=plt.subplots()
ax=sns.violinplot(x='time',y='total_bill',hue='sex',data=tips,split=True)
```

for문을 통해 스타일 적용

```python
fig=plt.figure()
seaborn_styles=['darkgrid','whitegrid','dark','white','ticks']
for idx, style in enumerate(seaborn_styles):
  plot_position=idx + 1
  with sns.axes_style(style):
    ax=fig.add_subplot(2,3,plot_position)
    violin=sns.violinplot(x='time',y='total_bill',data=tips,ax=ax)
violin.set_title(style)
fig.tight_layout()
```
